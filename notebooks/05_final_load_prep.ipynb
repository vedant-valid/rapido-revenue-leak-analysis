## 📂 Load Final Dataset

We load the cleaned dataset generated from previous steps  
(Extraction → Cleaning → EDA → Analysis).

```python
df = pd.read_csv("../data/processed/cleaned_rides.csv")

print(f"Final Dataset: {df.shape[0]:,} rows x {df.shape[1]} columns")
df.head()
```

## 🔍 Data Quality Checks

We validate:
- Missing values
- Distribution of ride status
- Data types

```python
print("=== NULL VALUES ===")
print(f"Total nulls: {df.isnull().sum().sum()}")

print("\n=== RIDE STATUS DISTRIBUTION ===")
print(df['ride_status'].value_counts(normalize=True) * 100)

print("\n=== DATA TYPES ===")
print(df.dtypes.value_counts())
```

## 📊 KPI Summary

We compute core business metrics required for dashboard:

- Total Rides
- Revenue
- Cancellation Rate
- Completion Rate
- Average Fare
- Distance & Duration

```python
total_rides = len(df)
total_revenue = df[df['completed']==1]['total_fare'].sum()
avg_fare = df[df['completed']==1]['total_fare'].mean()
completion_rate = df['completed'].mean()
cancel_rate =  1 - completion_rate
avg_distance = df['distance'].mean()
avg_duration = df['duration'].mean()

print("=== KPI SUMMARY ===")
print(f"Total Rides: {total_rides:,}")
print(f"Total Revenue: ₹{total_revenue:,.0f}")
print(f"Avg Fare: ₹{avg_fare:.2f}")
print(f"Cancellation Rate: {cancel_rate*100:.2f}%")
print(f"Completion Rate: {completion_rate*100:.2f}%")
print(f"Avg Distance: {avg_distance:.2f} km")
print(f"Avg Duration: {avg_duration:.2f} min")
```

## 💰 Business KPI — Revenue Loss

We estimate revenue lost due to ride cancellations.

This is a critical business metric for decision-making.

```python
total_cancelled = (df['completed'] == 0).sum()

lost_revenue = total_cancelled * avg_fare

print(f"Estimated Revenue Lost: ₹{lost_revenue:,.0f}")
```
